```python
def get_df_source12():

    def get_N():
        df = df_panel[df_panel['t']==1][['gvkey_hscode6', 'source1']].groupby(['gvkey_hscode6']).nunique()
        df.reset_index(inplace=True)
        df.rename(columns={'source1': 'N'}, inplace=True)
        dictio = {gvhs: n for gvhs, n in zip(df['gvkey_hscode6'], df['N'])}

        N = dict()
        N[1] = list(df['gvkey_hscode6'][df['N']==1].unique())
        N[2] = list(df['gvkey_hscode6'][df['N']==2].unique())
        N[3] = list(df['gvkey_hscode6'][df['N']==3].unique())
        N[0] = list(set(df_panel['gvkey_hscode6'].unique()).difference(set(N[1])).difference(set(N[2])).difference(set(N[3])))

        return N

    N = get_N()

    # Obs. con 2 sources distintas
    lista = []
    for gvkey in N[2]:
        df = df_panel[df_panel['gvkey_hscode6']==gvkey].copy()
        sources = list(df[df['t']==1]['source1'].unique())
        if len(sources)==1: continue
        lista.append(gvkey)

    df_panel2 = df_panel[np.logical_and(df_panel['gvkey_hscode6'].isin(N[2]), df_panel['t']==1)].copy()

    df = df_panel2[['gvkey_hscode6', 'source1', 'date']].groupby(['gvkey_hscode6', 'source1']).first()
    df.reset_index(inplace=True)
    df_source1 = df[['gvkey_hscode6', 'source1']].copy()
    df_source1

    from datetime import datetime
    from dateutil import relativedelta
    dictio = dict()
    for gv_hs in df_source1['gvkey_hscode6'].unique():
        country = list(df_source1[df_source1['gvkey_hscode6']==gv_hs]['source1'])
        dates1 = list(df_panel2[np.logical_and(df_panel2['gvkey_hscode6']==gv_hs, df_panel2['source1']==country[0])]['date'].copy())
        dates2 = list(df_panel2[np.logical_and(df_panel2['gvkey_hscode6']==gv_hs, df_panel2['source1']==country[1])]['date'].copy())
        if dates1[-1] < dates2[0]: dictio[gv_hs] = [country[0], country[1]]
        if dates1[0] > dates2[-1]: dictio[gv_hs] = [country[1], country[0]]

    df = pd.DataFrame()
    df['gvkey_hscode6'] = dictio.keys()
    df['source1'] = [lista[0] for lista in dictio.values()]
    df['source2'] = [lista[1] for lista in dictio.values()]
    return df
```

```python
df_source12 = get_df_source12()
df_source12
```

```python
# lista de las empresas con source shift desde China a Taiwan o HongKong
firms_china_hongkong_taiwan = df_source12[np.logical_and(df_source12['source1']=='China', df_source12['source2'].isin(['Hong Kong', 'Taiwan']))]['gvkey_hscode6']
```

```python
df_panel.columns
```

```python
# Una empresa que sale de China a HongKong o Taiwan, no sale realmente de China.
# Por lo tanto, no se considera un periodo de shift.
# T1=0 T2=0 china_inn=0 china_out=0

#  t: Son los 2 periodos de consistency.
# T1: 1er periodo de consistency para toda empresa que hace shift
# T2: 2do periodo de consistency para toda empresa que hace shift
# china_inn: Es el periodo que entra a China (por lo tanto, es China)
# china_out: Son los 2 periodos de consistency para una empresa que sale de China
```

```python
# T1=0 T2=0 china_inn=0 china_out=0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_china_hongkong_taiwan), 'T1'] = 0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_china_hongkong_taiwan), 'T2'] = 0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_china_hongkong_taiwan), 'china_inn'] = 0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_china_hongkong_taiwan), 'china_out'] = 0
```

```python
# lista de las empresas con source shift desde Taiwan/HongKong -----> China
firms_hongkong_taiwan_china = df_source12[np.logical_and(df_source12['source2']=='China', df_source12['source1'].isin(['Hong Kong', 'Taiwan']))]['gvkey_hscode6']
```

```python
# T1=0 T2=0 china_inn=0 china_out=0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_hongkong_taiwan_china), 'T1'] = 0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_hongkong_taiwan_china), 'T2'] = 0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_hongkong_taiwan_china), 'china_inn'] = 0
df_panel.loc[df_panel['gvkey_hscode6'].isin(firms_hongkong_taiwan_china), 'china_out'] = 0
```
